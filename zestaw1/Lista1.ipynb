Zaimplementuj szyfr Cezara. Napisz funkcję $\textit{Cezar(tekst, klucz)}$ tak aby jako argumenty przyjmowała tekst do zaszysfrowania oraz klucz a zwracała zaszyfrowany tekst. Funkcja ta dostając jako argumenty zaszyfrowany tekst oraz -klucz (klucz, ktorym zaszyfrowano tekst ze znakiem minus) będzie zwracała odszyfrowany tekst. Funkcja powinna przejść testy, czyli powinna 8 razy zwrócić $\textit{True}$.

Pamiętaj o polskich znakach. Szyfrujemy tylko litery, natomiast cyfry, znaki interpunkcyjne i białe znaki zostają bez zmian. Wielkie litery po zaszyfrowaniu są również wielkie. Użyj poniższego alfabetu.

```python
alfabet={'a':0,'ą':1, 'b':2, 'c':3, 'ć':4, 'd':5, 'e':6, 'ę':7, 'f':8, 'g':9, 'h':10, 'i':11, 'j':12, 'k':13, 'l':14, 'ł':15,
             'm':16, 'n':17, 'ń':18, 'o':19,'ó':20, 'p':21, 'q':22,'r':23, 's':24, 'ś':25, 't':26, 'u':27,'v':28, 'w':29, 'x':30,'y':31, 'z':32,'ź':33,'ż':34}
```

Przy implementacji mogą być przydatne takie elementy biblioteki string jak: $\textit{string.punctuation, string.whitespace, string.digits, upper(), lower(), isupper()}$.

```python
def Cezar(tekst,klucz):
    encrypted=''
    for c in tekst:
        new_c=''
        if(c in string.punctuation or c in string.whitespace or c in string.digits):
            new_c=c
        else:
            if(c.isupper()):
                i = (alfabet[c.lower()] + klucz) % 35
                new_c=list(alfabet.keys())[list(alfabet.values()).index(i)].upper()
            else:
                i = (alfabet[c] + klucz) % 35
                new_c=list(alfabet.keys())[list(alfabet.values()).index(i)]
        encrypted+=new_c
    return encrypted
```

Podobnie funkcja $\textit{Vignere_odszyfruj(tekst, klucz)}$ przyjmuje jako argumenty tekst zaszyfrowany oraz klucz w postaci słowa. Może być przydatne skorzystanie z zamiany liter na liczby (tzn. na liczby określające pozycję litery w alfabecie):

```python
liczb_do_lit=dict()
lit_do_liczb=dict()
for i,l in enumerate(alfabet):
    liczb_do_lit[i]=l
    lit_do_liczb[l]=i
```

```python
def Vignere_zaszyfruj(tekst,klucz):
    encrypted=''
    n = len(klucz)
    indeks=0
    for c in tekst:
        indeks%=n
        if(c in string.punctuation or c in string.whitespace or c in string.digits):
            new_c=c
        else:
            new_c=Cezar(c,alfabet[klucz[indeks]])
            indeks+=1
        encrypted+=new_c
    return encrypted 
```
